```python
# @title
# ============================================
# 1. Estrategia 1 – GPS (línea simple)
# ============================================
df1 = pd.DataFrame({"Año": [2021, 2022, 2023], "Puntaje": [50, 60, 80]})
fig, ax = plt.subplots(figsize=(4, 2.2))
sns.lineplot(data=df1, x="Año", y="Puntaje", marker="o", ax=ax)
ax.set_title("1. GPS – Evolución puntaje Matemáticas", pad=8)
ax.set_ylim(45, 85)
ax.set_ylabel("Puntaje")
plt.tight_layout()
plt.savefig("estrategia1_GPS.png", dpi=300)
plt.show()
```

Estrategia 1 – GPS de la gráfica
Gráfica: línea ascendente 2021-2023 (puntaje Matemáticas 50 → 80)

Análisis guiado

a) G (Guía) – ¿Qué muestra?

Variable Y = puntaje promedio Matemáticas; X = años.

b) P (Patrones) – ¿Qué veo?

Tendencia lineal positiva sin saltos bruscos.

c) S (Significado) – ¿Por qué importa?

El aumento es constante; probablemente el colegio implementó refuerzo.

d) Decisión de estudio

• Identificar qué temas subieron más y replicar esa estrategia en otras áreas.

• Planificar 2 horas semanales de práctica dirigida si mi puntaje está < 60.

```python
# ============================================
# 2. Estrategia 2 – Traffic-light (barras verticales coloreadas)
# ============================================
df2 = pd.DataFrame({"Año": [2020, 2021, 2022, 2023], "Brecha": [35, 30, 20, 15]})
colors = ["red" if b >= 25 else "orange" if b >= 20 else "green" for b in df2["Brecha"]]
fig, ax = plt.subplots(figsize=(4, 2.2))
sns.barplot(data=df2, x="Año", y="Brecha", palette=colors, ax=ax)
ax.set_title("2. Traffic-light – Brecha oficial-no oficial", pad=8)
ax.set_ylabel("Diferencia (puntos)")
plt.tight_layout()
plt.savefig("estrategia2_Traffic.png")
plt.show()

```

Estrategia 2 – Traffic-light
Gráfica: barras coloreadas de la brecha oficial-no oficial.

Análisis guiado

a) Colores rápidos

Rojo 2020-2021 → brecha crítica (≥ 25 puntos).

Verde 2023 → brecha mínima (15).

b) Pregunta clave

¿Qué cambió en 2022-2023? (Becas, acompañamiento, internet).

c) Decisión de estudio

• Si estudio en colegio no oficial con alto puntaje: compartir con compañeros oficiales los materiales usados.

• Si estudio en colegio oficial: buscar los programas que se aplicaron en 2022 y replicar sus ejercicios de razonamiento cuantitativo.

```python
# ============================================
# 3. Estrategia 3 – ANCLA (gráfico de puntos + barras doble eje)
# ============================================
df3 = pd.DataFrame({
    "Grupo": ["Con internet", "Sin internet"],
    "Pct": [78, 32],
    "Score": [260, 210]
})
fig, ax1 = plt.subplots(figsize=(4, 2.2))
bars = ax1.bar(df3["Grupo"], df3["Pct"], color=["skyblue", "salmon"])
ax1.set_ylabel("% estudiantes", color="black")
ax1.tick_params(axis='y', labelcolor="black")
ax2 = ax1.twinx()
ax2.plot(df3["Grupo"], df3["Score"], color="navy", marker="o", linewidth=2)
ax2.set_ylabel("Puntaje promedio", color="navy")
ax2.tick_params(axis='y', labelcolor="navy")
ax1.set_title("3. ANCLA – Acceso vs Puntaje", pad=8)
plt.tight_layout()
plt.savefig("estrategia3_ANCLA.png")
plt.show()

```

Estrategia 3 – ANCLA

Gráfica: doble eje – % acceso a internet vs puntaje.

Análisis guiado

A – Ejes claros.

N – Dos series (con/sin internet).

C – Diferencia 23,8 %.

L – Muestra urbana (cuidado al inferir rural).

A – Aplicación: cada 10 % más de acceso se asocia a ~8 puntos más.

Decisión de estudio

• Crear horario fijo de 30 min diarios usando plataformas gratuitas (Khan, Coursera) que requieran conexión.

```python

# ============================================
# 6. Estrategia 6 – Paréntesis mental (área / stack)
# ============================================
years = [2021, 2022, 2023]
mate = [50, 60, 80]
ingl = [45, 50, 55]
cie  = [60, 65, 70]
df6 = pd.DataFrame({"Año": years, "Matemáticas": mate, "Inglés": ingl, "Ciencias": cie})
fig, ax = plt.subplots(figsize=(4, 2.2))
ax.stackplot(df6["Año"], df6["Matemáticas"], df6["Inglés"], df6["Ciencias"],
             labels=["Matemáticas", "Inglés", "Ciencias"],
             colors=["#ffd166", "#ef476f", "#118ab2"])
ax.legend(loc="upper left", bbox_to_anchor=(1,1))
ax.set_title("6. Paréntesis mental – Áreas apiladas", pad=8)
ax.set_ylabel("Puntaje")
plt.tight_layout()
plt.savefig("estrategia6_Parentesis.png")
plt.show()
```

Estrategia 6 – Paréntesis mental

```python

# ============================================
# 7. Estrategia 7 – Fórmula rápida (bullet chart meta)
# ============================================
fig, ax = plt.subplots(figsize=(4, 1.2))
ax.barh(0, 220, color="lightgray", height=0.3, label="Simulacro inicial")
ax.barh(0, 240, color="seagreen", height=0.3, label="Meta +8.98 %")
ax.axvline(245, color="black", linestyle="--", linewidth=1)
ax.axvline(267, color="black", linestyle="-", linewidth=1)
ax.set_yticks([0])
ax.set_yticklabels(["Puntaje"])
ax.set_title("7. Meta +8.98 % = 240 puntos", pad=8)
ax.legend(loc="center right")
ax.set_xlim(200, 270)
plt.tight_layout()
plt.savefig("estrategia7_Formula.png")
plt.show()

```

Estrategia 7 – Mapa de fórmulas

Gráfica: línea con variación 245 → 267 (+8,98 %).

Análisis guiado

a) Fórmula rápida: (267-245)/245*100.

b) Meta personal: si mi último simulacro fue 220, ¿cuánto necesito para +8,98 %?
220×1,0898 ≈ 240 → meta 240 puntos.

Decisión de estudio

```python
# ============================================
# 8. Estrategia 8 – Feedback pares (dispersión + regresión)
# ============================================
hrs = [0,1,2,3,4,5]
scr = [190,210,235,260,270,275]
fig, ax = plt.subplots(figsize=(4, 2.2))
sns.scatterplot(x=hrs, y=scr, ax=ax, color="navy")
sns.regplot(x=hrs, y=scr, ax=ax, scatter=False, color="crimson")
ax.set_title("8. Dispersión – Horas vs Puntaje", pad=8)
ax.set_xlabel("Horas estudio/semana")
ax.set_ylabel("Puntaje")
plt.tight_layout()
plt.savefig("estrategia8_Dispersion.png")
plt.show()
```

Estrategia 8 – Feedback pares

Gráfica: dispersión Horas vs Puntaje.

Análisis guiado

a) Regresión muestra que después de 3 h/semana las ganancias se desaceleran.

```python

# ============================================
# 10. Estrategia 10 – Datos DANE (líneas suavizadas)
# ============================================
estrato = np.array([1,2,3,4,5,6])
internet = np.array([42,58,72,85,93,97])
fig, ax = plt.subplots(figsize=(4, 2.2))
sns.lineplot(x=estrato, y=internet, marker="o", ax=ax, color="green")
ax.fill_between(estrato, internet, alpha=0.2, color="green")
ax.set_title("10. Acceso a internet por estrato", pad=8)
ax.set_xlabel("Estrato")
ax.set_ylabel("% hogares")
ax.set_ylim(0,100)
plt.tight_layout()
plt.savefig("estrategia10_DANE.png")
plt.show()
```

Estrategia 10 – Datos DANE / GeoGebra

Gráfica: línea % acceso internet por estrato.

Análisis guiado

a) Pendiente fuerte entre estrato 1 y 3, luego se estabiliza.

b) Interpretación: intervenciones en estratos 1-2 tendrían mayor impacto.
